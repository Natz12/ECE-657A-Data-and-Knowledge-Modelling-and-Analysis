## Homework1 Data Summarization

The Homework aims to use the Breast Cancer Wisconsin (Diagnostic) Data Set to summarize single variable data through the computation of some basic statistics and plots. Python 3.7.1 was used for this purpose.

The wdbc.data [1] file was downloaded from the link https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic), and the file wdbc.names [2] containing the data set documentation was used to understand its contents and to identify the column labels.

The pandas [3] and Matplotlib [4] toolkit documentation was used as a guide to solve the problems presented in Homework 1.
Previously taken DataCamp [5] courses were also helpful in solving the homework.

### Libraries used
Pandas was imported to handle the Dataframes and to make the statistics calculations (mean, standard deviation, variance, skewness and mode).

Matplotlib.pyplot was used to plot the histograms and to enhance the readability and visualization.

```python
##--Importing necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

### Header Creation
The dataset was inspected and it was noticed that it didn't contain a header or column names. The header was manually created. Notice that the column names were created according to the specification in [2].

```python
##---Naming the columns
#Column name parts
column1 = ['Radius', 'Texture', 'Perimeter', 'Area', 'Smoothness', 'Compactness', 'Concavity', 'Concave_points', 'Symmetry', 'Fractal_dimension']
column2 = ['Mean', 'SE', 'Worst']
column_list = list(range(32))
column_list[0] = 'ID number'
column_list[1] = 'Diagnosis'

#getting the columns names according to the documentation (wdbc.names)
for x in range(10):
    column_list[x+2] = column2[0] + '_' + column1[x]
    column_list[x+12] = column1[x] + '_' + column2[1] #Order changes to follow the documentation's specifications
    column_list[x+22] = column2[2] + '_' + column1[x]
```

### Reading the file
The file was taken directly from the dataset url and assigned its corresponding column names. The first 4 rows can be seen in the table below.


```python
##--Reading the file

#From URL
file='https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'

#Read the file
df = pd.read_csv(file, header = None, names = column_list)
df.head(4)
```

The dataset represents the features computed from a digitalized image of cell nuclei from samples of a breast mass. The dataset consists of 569 entries and 32 attributes/columns. The samples were collected on november 1995 and classified between benign and malignant according to their diagnosis. The data type of each attribute can be observed in the information below.

```python
df.info()
```

The columns are comprised of the ID number of the patient stored as an int, the diagnosis (M for malignant and B for benign) represented as an object and 30 features stored as floats as shown within the previous code. None of the 32 attributes have missing values.

### Summary statistics
Summary statistics help to simplify the description of a large data set. They make more manageable the task of representing the basic features of the samples [7]. 

The mean, standard deviation, variance, mode and skewness of the dataset were calculated using methods included in the pandas dataframe structure.

There are many ways to calculate the mode of a sample, but most of them return an array, Dataframe or tuple, either with one of the repetitions with the number of times the value is present in the sample or all the results that appear the same number of times as seen in the following table.

```python
##---Statistics
#Selecting the columns of interest
df_describe = df.iloc[:,2:]

#Calculating the stats
df_mode = df_describe.mode()
df_mode
```

During this exercise, the mode selected will be the smallest value with the most repetitions, that is, if many values appear the same number of times only the smallest value will appear in the result. This value is taken from the first row of the array shown above.

```python
df_mode = df_mode.iloc[0,:]
df_mean = df_describe.mean()
df_std = df_describe.std()
df_var = df_describe.var()
df_skew = df_describe.skew()

#Putting all together
df_results = pd.concat([df_mean, df_mode, df_skew, df_std, df_var],axis=1,keys=["Mean","Mode","Skew","Std","Var"])
df_results
```

The dataframe df_results shows the statistical variables of the dataset. The mean tries to describe the center of the data but since the skewness of every feature is different from zero we can conclude that the distribution is not symmetrical and the mean might not be the best statistic value to describe the data [8].  

## Pearson Correlation Coefficient (PCC)
The PCC is used to recognize the degree of association between two variables and looks for dependency strength. The advantage of the PCC is that it is a descriptive measure of association that doesn't depend on the scale of measurement of the variables of study [6].

Using python, the PCC could be obtained using scipy.stats.pearsonr, but this function only accepts 2 arrays making the correlation within one dataframe more difficult. Pandas' correlation method offers this feature and with one line of code it shows us the correlation of all the combinations of columns of the Dataframe.

```python
##--Pearson Coefficient
pear_corr = df_describe.corr(method = 'pearson')

#Visualizing the first 8 columns
pear_corr.iloc[:,:8]
```

The first thing to notice is that, as expected, the correlation between the Radius, Perimeter and Area are high and close to 1, signaling a positive dependency. This is expected as the radius, perimeter and area are geometrically linked and their dependency was already known.

Another noticeable high correlation coefficient is the one between the mean radius and the mean concave points, showing that when the mean radius increases, so does the mean concave points. There is also a positive correlation between the concavity, concave point and compactness.

The worst perimeter and fractal dimension SE, which is not visible int he table above, presented a correlation of -0.001. This small correlation indicates that there is low or none existent linear relationship between the two features.

The strongest negative correlation was between mean radius and mean fractal dimension showing that when the radius increases, the fractal dimension decreases. Even though this was the strongest negative correlation, a correlation of -0.31 is not considered very strong. [6] 


## Histograms
Histograms represent continuous data graphically by dividing the range of the observations into bins and representing the number of times each value division occurred. They are helpful to analyze frequencies and trends. With them it is possible to visualize useful information such as the minimum and maximum frequency and the data distribution [9].

```python
##--Histograms
#kind selects the type of plot

df_describe.Mean_Radius.plot(kind='hist')
plt.legend()
plt.title("Tumor Cells Mean Radius")

plt.show()
```

The histogram of the mean radius shows a skewed distribution with a long tail to the right. It can also be confirmed that the mode is located around 12.

Sometimes histograms are also useful to compare distributions between categories when they are plotted together.

```python
##--Histograms
#kind selects the type of plot
#color selects the color of the plot
#alpha indicates the transparency
#Label specifies the name of the data
df_describe.Mean_Radius[df.Diagnosis=='M'].plot(kind='hist', color='r', alpha=0.5, label='M')
df_describe.Mean_Radius[df.Diagnosis=='B'].plot(kind='hist', color='g', alpha=0.5, label='B')
plt.legend()
plt.xlabel("Mean_Radius")
plt.title("Comparison of Malignant and Benign Tumor Cells' Mean Radius")

plt.show()

df_describe.Worst_Radius[df.Diagnosis=='M'].plot(kind='hist', color='r', alpha=0.5, label='M')
df_describe.Worst_Radius[df.Diagnosis=='B'].plot(kind='hist', color='g', alpha=0.5, label='B')
plt.legend()
plt.xlabel("Worst_Radius")
plt.title("Comparison of Malignant and Benign Tumor Cells' Worst Radius")

plt.show()

df_describe.Mean_Fractal_dimension[df.Diagnosis=='M'].plot(kind='hist', color='r', alpha=0.5, label='M')
df_describe.Mean_Fractal_dimension[df.Diagnosis=='B'].plot(kind='hist', color='g', alpha=0.5, label='B')
plt.legend()
plt.xlabel("Mean_Fractal_dimension")
plt.title("Comparison of Malignant and Benign Tumor Cells' Mean Fractal Dimension")

plt.show()
```

The first histogram shows the distribution of the malignant and benign tumor cells' mean radius classifying the samples by their diagnosis. This histogram shows two distinct distributions that, although overlapping in the intermediate values, have a visible trend that separates them. The same thing happens with the second histogram which compares the worst radius of the two diagnoses and also has a visible separation. 

Nevertheless, not all histograms will provide simple visual trends. For example, the third histogram doesn’t show a visually perceptible division between the mean smoothness for the two diagnoses.
